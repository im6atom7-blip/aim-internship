```python
def power (base, exponent=2):
    """Raises base to the power of exponent (default is 2)"""
    return base** exponent
# using default parameter
print(power(51))    #25(5^2)
print(power(2,3))  #8(2^3)
print(power(3,4))  #81(3^4)  
```

```python
def circle_calculation(radius):
    """calculates area and circumference of a circle"""
    import math
    area = math.pi*radius **2
    circumference = 2*math.pi*radius
    return area, circumference
# unpacking returned values
area,circ=circle_calculation(7)
print(f"Area:{area:.2f}")
print(f"circumference:{circ:.2f}")
```

```python
def show_args(*args):
    print("args:",args)
def show_kwargs(**kwargs):
    print("kwargs:",kwargs)
show_args(1,2,3)
show_kwargs(name="Bob",age=25)
```

```python
def sum_numbers(*args):
    """sum any number of arguments"""
    total  =0
    for num in args:
        total+= num
    return total
#using variable arguments
print(sum_numbers(1,2,3))
print(sum_numbers(10,20,30,40))
print(sum_numbers(5))
```

```python
def create_person(**kwargs):
    """creates a person dictionary with given attributes"""
    person ={}
    for key,value,in kwargs.items():
        person[key]=value
    return person
#using keyword arguments
person1 = create_person(name="alice",age=25,city="new work")
person2 = create_person(name="Bob",occupation="engineer")
print(person1)
print(person2)
```

```python
#example 1: factorial using recursion
def factorial(n):
    if n==0 or n ==1:   #base case
        return 1
    else:               #recursive case
        return n*factorial(n-1)
print(factorial(5))     #120
        
```

```python
#example 2: Fibonacci using recursion
def fibonacci(n):
    if n==0: #base case
        return 0
    elif n==1:#base case
        return 1
    else: #recursive case
        return fibonacci(n-1)+fibonacci(n-2)
print(fibonacci(6)) #8
```

```python
#example 3: recursive sum of a lst
def recursive_sum(lst):
    if not lst: #base case(empty list)
        return 0
    else:       #recursive case
        return lst[0]+recursive_sum(lst[1:])
print(recursive_sum([1,2,3,4,5])) #15
```

```python
def calculate_discount(price: float, discount_rate: float = 0.10) -> float:
 """
 Calculate the final price after applying a discount.
 This function computes the discounted price of a product or service.
 It validates inputs to ensure the price is non-negative and the discount
 rate is between 0 and 1 (inclusive). If inputs are invalid, a ValueError
 is raised.
 Args:
 price (float): Original price before discount.
 discount_rate (float, optional): The discount rate to apply.
 Must be between 0 and 1. Defaults to 0.10 (10%).
 Returns:
 float: Final price after discount is applied.
 Raises:
 ValueError: If `price` is negative.
 ValueError: If `discount_rate` is not between 0 and 1.
 Example:
 >>> calculate_discount(100.0, 0.20)
 80.0
 >>> calculate_discount(250.0) # Uses default discount of 10%
 225.0
 """
 if price < 0:
     raise ValueError("Price cannot be negative")
 if not (0 <= discount_rate <= 1):
     raise ValueError("Discount rate must be between 0 and 1")

 return price * (1 - discount_rate)
# Example usage
final_price = calculate_discount(1200, 0.25)
print(f"Final price after discount: ${final_price:,.2f}")
```
