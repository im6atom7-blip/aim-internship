```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0

    for char in text:
        if char in vowels:
            count += 1

    return count

print(count_vowels("Python"))
```

```python
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

print(is_prime(7))
```

```python
def fibonacci(n):
    series = []
    a, b = 0, 1

    for _ in range(n):
        series.append(a)
        a, b = b, a + b

    return series

print(fibonacci(6))
```

```python
def sum_of_digits(num):
    total = 0

    while num > 0:
        total += num % 10
        num //= 10

    return total

print(sum_of_digits(1234))
```

```python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True

    return False

print(is_leap_year(2024))
```
