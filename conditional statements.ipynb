```python
#checking eligibility for voting and citizenship
age = 20
citizen = "no"
if age >= 18:
    if citizen == "yes":
        print("You are eligible to vote")
    else:
        print("You must be a citizen to vote")
else:
    print("You must be 18 or older to vote")
```
