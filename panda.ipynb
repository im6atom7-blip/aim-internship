```python
import pandas as pd
```

```python
# From dictionary
data = {'Name': ['Alice', 'Bob'], 'Age': [25, 30]}
df = pd.DataFrame(data)
```

```python
# From list of lists
lst = [['Alice', 25], ['Bob', 30]]
df = pd.DataFrame(lst, columns=['Name', 'Age'])
```

```python
df.head() # First 5 rows
df.tail() # Last 5 rows
df.info() # Info about DataFrame
df.describe() # Summary statistics
```

```python
# Column selection
df['Name']
df[['Name', 'Age']]
```

```python
# Row selection by label and position

df.loc[0]
df.iloc[1]
```

```python
df[df['Age'] > 25] # Rows where Age > 25
```

```python
import pandas as pd

try:
    df['Gender'] = ['F', 'M']  # Add new column
except NameError:
    df = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30], 'Gender': ['F', 'M']})
df.drop('Age', axis=1, inplace=True) # Remove column
```

```python
df.isnull()
df.dropna()
df.fillna(0)
```

```python
df.groupby('Gender')['Age'].mean()
```
